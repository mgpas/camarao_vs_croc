```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, collect_list, monotonically_increasing_id

# Inicializar a sessão do Spark
spark = SparkSession.builder \
    .appName(" [Gold] LOL Croc Match Data Lake") \
    .getOrCreate()

# Diretório contendo os dados processados na camada Silver
crocData_dir = "/content/data_lake/silver/crocConsolidated"

# Ler os dados da camada Silver
croc_df = spark.read.parquet(crocData_dir)
croc_df = croc_df.coalesce(1)

# Filtrar os dados pelo PUUID específico
filtered_df = croc_df.filter(F.col("puuid") == crocPUUID)

rounded_df = filtered_df.withColumn("damagePerMinute", round(col("damagePerMinute"), 3))\
.withColumn("gameLength", round(col("gameLength"), 3))\
.withColumn("teamDamagePercentage", round(col("teamDamagePercentage"), 3))\
.withColumn("kda", round(col("kda"), 3))

crocFiltered_df = rounded_df.withColumn("temp_id", monotonically_increasing_id())

# Exibir todos os registros do DataFrame de dados processados
print(f"Dados processados (Camada Gold):")
crocFiltered_df.show()

# Criar o diretório gold se ele não existir
gold_dir = "data_lake/gold/"
if not os.path.exists(gold_dir):
    os.makedirs(gold_dir)

# Escrever os dados consolidados em um diretório silver
crocFiltered_dir = "data_lake/gold/crocFiltered/"
crocFiltered_df.write.mode("overwrite").parquet(crocFiltered_dir)

# Define a custom CSV filename (change as needed)
csv_filename = "croc_filtered.csv"

crocFiltered_df.write.mode("overwrite")\
    .option("header", True) \
    .csv(os.path.join(gold_dir, csv_filename))

json_filename = "croc_filtered.json"
crocFiltered_df.write.mode("overwrite")\
    .json(os.path.join(gold_dir, json_filename))

# Encerrar a sessão do Spark
spark.stop()
```
